```python
mars_df.count()
```

```python
# 2. How many Martian days' worth of data are there?
```

```python
avg_monthly_temp = mars_df.groupby('month').mean().loc[:,'min_temp']
avg_monthly_temp
```

```python
# 3. What is the average low temperature by month?
```

```python
avg_monthly_temp_df = pd.DataFrame(avg_monthly_temp)
avg_monthly_temp_df
```

```python
Temps=avg_monthly_temp_df['min_temp']
Temps
```

```python
months=avg_monthly_temp_df.index
months
```

```python
plt.bar(months,Temps)
plt.xlabel("month")
plt.ylabel("Temperature in Celsius")
```

```python
# Plot the average temperature by month
```

```python
sorted_avg_monthly_temp_df = avg_monthly_temp_df.sort_values(by=['min_temp'])
sorted_avg_monthly_temp_df
```

```python
sorted_temps=sorted_avg_monthly_temp_df['min_temp']
sorted_months=sorted_avg_monthly_temp_df.index
sorted_temps
```

```python
#sorted bar chart in desc order
plt.bar(range(len(sorted_temps)),sorted_temps)
plt.xticks(range(len(sorted_temps)),sorted_months)
plt.xlabel("month")
plt.ylabel("Temperature in Celsius")
```

```python
# Identify the coldest and hottest months in Curiosity's location
```

```python
avg_monthly_pressure=mars_df.groupby('month').mean().loc[:,['pressure']]
avg_monthly_pressure
```

```python
avg_monthly_pressure_sorted=avg_monthly_pressure.sort_values(by='pressure')
avg_monthly_pressure_sorted
```

```python
pressure=avg_monthly_pressure_sorted['pressure']
months=avg_monthly_pressure_sorted.index
```

```python
# 4. Average pressure by Martian month
```

```python
plt.bar(range(len(pressure)),pressure)
plt.xticks(range(len(pressure)),months)
plt.xlabel("month")
plt.ylabel("Atmospheric Pressure")
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
